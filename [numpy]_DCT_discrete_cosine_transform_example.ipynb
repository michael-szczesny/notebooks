```
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```
x = np.linspace(-2*np.pi, 2*np.pi, 200)
```

```
def f(x):
    return .8*np.cos(7*x)+1.2*np.cos(4*x)+.5*np.random.randn(len(x))
```

```
plt.figure(figsize=(10,5))
dft = []
for k in range(1,10):
    y = np.cos(k*x)
    d = np.trapz(y*f(x))/(.5*len(x))
    if d > .4:
        plt.plot(x,y)
    dft.append(d)
plt.plot(x,f(x), 'r--', alpha=.4);
```

```
plt.plot(range(1,10),dft,'o');
```
